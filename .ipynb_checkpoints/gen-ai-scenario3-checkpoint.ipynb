Task: Create a SQL query to identify:

* User preference trends for the identified customers

# Generated Code:
`-- Customer Segmentation Query for High-Value Active Users
-- Purpose: Identify top 20% spending customers who are active for feature rollout

WITH active_users AS (
    -- Step 1: Identify users active in last 30 days
    SELECT DISTINCT user_id
    FROM user_activity 
    WHERE last_login_date >= CURRENT_DATE - INTERVAL '30 days'
),

user_spending AS (
    -- Step 2: Calculate total spending per user
    SELECT 
        t.user_id,
        SUM(t.amount) as total_spending,
        COUNT(t.transaction_id) as transaction_count,
        MAX(t.transaction_date) as last_transaction_date,
        MIN(t.transaction_date) as first_transaction_date,
        COUNT(DISTINCT t.platform) as platform_count
    FROM transactions t
    WHERE t.user_id IN (SELECT user_id FROM active_users)
    GROUP BY t.user_id
),

spending_percentile AS (
    -- Step 3: Calculate spending percentiles to identify top 20%
    SELECT 
        user_id,
        total_spending,
        transaction_count,
        last_transaction_date,
        first_transaction_date,
        platform_count,
        PERCENT_RANK() OVER (ORDER BY total_spending) as spending_percentile
    FROM user_spending
),

high_value_users AS (
    -- Step 4: Filter for top 20% spenders
    SELECT user_id, total_spending, transaction_count, 
           last_transaction_date, first_transaction_date, 
           platform_count, spending_percentile
    FROM spending_percentile 
    WHERE spending_percentile >= 0.8
),

user_segments AS (
    -- Step 5: Combine user data with activity metrics
    SELECT 
        hvu.user_id,
        hvu.total_spending,
        hvu.transaction_count,
        hvu.last_transaction_date,
        hvu.first_transaction_date,
        hvu.platform_count,
        ua.last_login_date,
        ua.feature_usage_count,
        ua.account_type,
        CASE 
            WHEN hvu.total_spending >= 1000 AND hvu.transaction_count >= 10 THEN 'Premium'
            WHEN hvu.total_spending >= 500 THEN 'High Value'
            ELSE 'Standard High'
        END as customer_tier,
        CASE 
            WHEN ua.last_login_date >= CURRENT_DATE - INTERVAL '7 days' THEN 'Highly Active'
            WHEN ua.last_login_date >= CURRENT_DATE - INTERVAL '14 days' THEN 'Active'
            ELSE 'Recently Active'
        END as activity_level
    FROM high_value_users hvu
    JOIN user_activity ua ON hvu.user_id = ua.user_id
)

-- Final Query: High-value active customers with preferences and trends
SELECT 
    us.user_id,
    us.customer_tier,
    us.activity_level,
    us.account_type,
    us.total_spending,
    us.transaction_count,
    us.platform_count,
    us.feature_usage_count,

LEFT JOIN user_preferences up ON us.user_id = up.user_id

````python

# Final Solution:
-- OPTIMIZED VERSION: Reduces CTEs and improves performance
-- Key optimizations: Early filtering, reduced data movement, simplified logic

WITH active_user_spending AS (
    -- Combine Steps 1-2: Filter and calculate spending in one pass
    SELECT 
        t.user_id,
        SUM(t.amount) as total_spending,
        COUNT(t.transaction_id) as transaction_count,
        MAX(t.transaction_date) as last_transaction_date,
        MIN(t.transaction_date) as first_transaction_date,
        COUNT(DISTINCT t.platform) as platform_count
    FROM transactions t
    WHERE EXISTS (
        SELECT 1 FROM user_activity ua 
        WHERE ua.user_id = t.user_id 
        AND ua.last_login_date >= CURRENT_DATE - INTERVAL '30 days'
    )
    GROUP BY t.user_id
),

high_value_active_users AS (
    -- Calculate percentile and filter in one step using window function filter
    SELECT 
        user_id,
        total_spending,
        transaction_count,
        last_transaction_date,
        first_transaction_date,
        platform_count,
        -- Calculate customer tier directly
        CASE 
            WHEN total_spending >= 1000 AND transaction_count >= 10 THEN 'Premium'
            WHEN total_spending >= 500 THEN 'High Value'
            ELSE 'Standard High'
        END as customer_tier
    FROM (
        SELECT *,
               PERCENT_RANK() OVER (ORDER BY total_spending) as spending_percentile
        FROM active_user_spending
    ) ranked
    WHERE spending_percentile >= 0.8
)

-- Main query with direct joins - no intermediate user_segments CTE
SELECT 
    hvu.user_id,
    hvu.customer_tier,
    
    -- Activity level calculated directly
    CASE 
        WHEN ua.last_login_date >= CURRENT_DATE - INTERVAL '7 days' THEN 'Highly Active'
        WHEN ua.last_login_date >= CURRENT_DATE - INTERVAL '14 days' THEN 'Active'
        ELSE 'Recently Active'
    END as activity_level,
    
    ua.account_type,
    hvu.total_spending,
    hvu.transaction_count,
    hvu.platform_count,
    ua.feature_usage_count,
    ua.last_login_date,
    hvu.last_transaction_date,
    
    -- User preferences
    up.communication_preference,
    up.interface_theme,
    up.notification_settings,
    
    -- Calculated metrics
    ROUND(hvu.total_spending / NULLIF(hvu.transaction_count, 0), 2) as avg_transaction_amount,
    EXTRACT(days FROM (hvu.last_transaction_date - hvu.first_transaction_date)) as customer_lifetime_days,
    
    -- Simplified engagement score
    ROUND(
        ua.feature_usage_count * 0.3 + 
        hvu.transaction_count * 0.4 + 
        CASE WHEN ua.last_login_date >= CURRENT_DATE - INTERVAL '7 days' THEN 20 ELSE 0 END * 0.3, 
        2
    ) as engagement_score

FROM high_value_active_users hvu
JOIN user_activity ua ON hvu.user_id = ua.user_id
LEFT JOIN user_preferences up ON hvu.user_id = up.user_id

ORDER BY hvu.total_spending DESC, ua.feature_usage_count DESC;

-- ALTERNATIVE APPROACH: Single query without CTEs (most performant for smaller datasets)
/*
SELECT 
    t.user_id,
    SUM(t.amount) as total_spending,
    COUNT(t.transaction_id) as transaction_count,
    ua.last_login_date,
    ua.feature_usage_count,
    ua.account_type,
    up.communication_preference,
    up.interface_theme,
    up.notification_settings
FROM transactions t
JOIN user_activity ua ON t.user_id = ua.user_id
LEFT JOIN user_preferences up ON t.user_id = up.user_id
WHERE ua.last_login_date >= CURRENT_DATE - INTERVAL '30 days'
  AND t.user_id IN (
      SELECT user_id 
      FROM (
          SELECT user_id, 
                 PERCENT_RANK() OVER (ORDER BY SUM(amount)) as pct_rank
          FROM transactions 
          GROUP BY user_id
      ) ranked
      WHERE pct_rank >= 0.8
  )
GROUP BY t.user_id, ua.last_login_date, ua.feature_usage_count, 
         ua.account_type, up.communication_preference, 
         up.interface_theme, up.notification_settings
ORDER BY total_spending DESC;
*/
````
