```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, Tuple, Any
import warnings

"""
CRITICAL BLINDSPOTS TO BE AWARE OF:

1. Session Definition Assumptions - PARTIALLY ADDRESSED
   - Risk: Code assumes session_id represents valid user sessions
   - Current fix: Added session duration validation and outlier detection
   - Remaining risk: Multi-day sessions or session gaps not fully handled

2. Data Type and Scale Issues - NOT ADDRESSED
   - Risk: Large datasets could cause memory issues with groupby operations
   - Risk: Integer overflow with very large duration values
   - Recommendation: Consider chunked processing for large datasets

3. Business Logic Edge Cases - NOT ADDRESSED
   - Risk: No distinction between meaningful engagement vs bot traffic
   - Risk: Page view definition unclear (reloads, API calls, etc.)
   - Recommendation: Add user agent filtering and engagement quality metrics

4. Time Zone and Temporal Issues - NOT ADDRESSED
   - Risk: Duration calculations don't account for time zones or DST
   - Risk: No validation for chronological order within sessions
   - Recommendation: Add timezone awareness and temporal validation

5. Data Cleaning Aggressiveness - PARTIALLY ADDRESSED
   - Risk: Overly aggressive cleaning masks data quality issues
   - Current approach: Added reporting of cleaning actions
   - Recommendation: Implement data quality scoring and alerting

6. Visualization Limitations - PARTIALLY ADDRESSED
   - Risk: Charts don't handle extreme outliers well
   - Current fix: Added outlier detection and capping for visualizations
   - Recommendation: Consider log scales for extreme distributions

7. Concurrent Data Issues - NOT ADDRESSED
   - Risk: Inconsistent results if data updated during analysis
   - Recommendation: Implement data snapshots or versioning
"""

def detect_outliers_and_validate_sessions(df_clean: pd.DataFrame) -> pd.DataFrame:
    """
    Detect outliers and validate session integrity - addresses critical blindspot.
    
    Args:
        df_clean: Cleaned DataFrame with basic validation applied
        
    Returns:
        DataFrame with outlier handling and session validation
    """
    print("\nSession integrity and outlier analysis:")
    
    # Group by session to analyze session-level metrics
    session_stats = df_clean.groupby('session_id').agg({
        'duration': 'sum',
        'page_views': 'sum'
    }).reset_index()
    
    # Define realistic thresholds for session validation
    MAX_REASONABLE_SESSION_DURATION = 4 * 3600  # 4 hours in seconds
    MAX_REASONABLE_PAGE_VIEWS = 200  # Maximum pages per session
    MIN_MEANINGFUL_DURATION = 1  # 1 second minimum
    
    # Identify outlier sessions
    duration_outliers = session_stats['duration'] > MAX_REASONABLE_SESSION_DURATION
    page_view_outliers = session_stats['page_views'] > MAX_REASONABLE_PAGE_VIEWS
    suspicious_duration = session_stats['duration'] < MIN_MEANINGFUL_DURATION
    
    print(f"Sessions with unrealistic duration (>{MAX_REASONABLE_SESSION_DURATION/3600:.1f}h): {duration_outliers.sum()}")
    print(f"Sessions with excessive page views (>{MAX_REASONABLE_PAGE_VIEWS}): {page_view_outliers.sum()}")
    print(f"Sessions with suspicious duration (<{MIN_MEANINGFUL_DURATION}s): {suspicious_duration.sum()}")
    
    # Cap outlier values instead of removing (preserves data while fixing metrics)
    outlier_sessions = session_stats[duration_outliers | page_view_outliers]['session_id']
    
    if len(outlier_sessions) > 0:
        print(f"Capping {len(outlier_sessions)} outlier sessions to reasonable limits")
        
        # Create session-level caps and apply back to original data
        for session_id in outlier_sessions:
            session_mask = df_clean['session_id'] == session_id
            session_data = df_clean[session_mask].copy()
            
            # Cap total session duration proportionally across all records
            total_duration = session_data['duration'].sum()
            if total_duration > MAX_REASONABLE_SESSION_DURATION:
                scaling_factor = MAX_REASONABLE_SESSION_DURATION / total_duration
                df_clean.loc[session_mask, 'duration'] = df_clean.loc[session_mask, 'duration'] * scaling_factor
            
            # Cap total page views proportionally
            total_page_views = session_data['page_views'].sum()
            if total_page_views > MAX_REASONABLE_PAGE_VIEWS:
                scaling_factor = MAX_REASONABLE_PAGE_VIEWS / total_page_views
                df_clean.loc[session_mask, 'page_views'] = np.maximum(
                    1, np.round(df_clean.loc[session_mask, 'page_views'] * scaling_factor)
                )
    
    # Flag suspicious sessions for further review
    suspicious_sessions = session_stats[suspicious_duration]['session_id']
    if len(suspicious_sessions) > 0:
        print(f"Warning: {len(suspicious_sessions)} sessions have very short durations - possible bot traffic")
    
    return df_clean

def validate_data(logs_df: pd.DataFrame) -> pd.DataFrame:
    """
    Validate and clean the input data.
    
    Args:
        logs_df: DataFrame with user engagement data
        
    Returns:
        Cleaned DataFrame
        
    Raises:
        ValueError: If required columns are missing or data is invalid
    """
    required_columns = ['session_id', 'duration', 'page_views', 'device_type']
    
    # Check for required columns
    missing_cols = [col for col in required_columns if col not in logs_df.columns]
    if missing_cols:
        raise ValueError(f"Missing required columns: {missing_cols}")
    
    if logs_df.empty:
        raise ValueError("DataFrame is empty")
    
    # Create a copy to avoid modifying original data
    df_clean = logs_df.copy()
    
    # Validate and clean data
    print("Data validation report:")
    print(f"Original rows: {len(df_clean)}")
    
    # Remove rows with null session_id
    null_sessions = df_clean['session_id'].isnull().sum()
    if null_sessions > 0:
        print(f"Removing {null_sessions} rows with null session_id")
        df_clean = df_clean.dropna(subset=['session_id'])
    
    # Handle negative durations
    negative_duration = (df_clean['duration'] < 0).sum()
    if negative_duration > 0:
        print(f"Found {negative_duration} rows with negative duration - setting to 0")
        df_clean.loc[df_clean['duration'] < 0, 'duration'] = 0
    
    # Handle invalid page_views (should be positive integers)
    invalid_page_views = (df_clean['page_views'] <= 0).sum()
    if invalid_page_views > 0:
        print(f"Found {invalid_page_views} rows with invalid page_views - setting minimum to 1")
        df_clean.loc[df_clean['page_views'] <= 0, 'page_views'] = 1
    
    # Handle missing device_type
    missing_device = df_clean['device_type'].isnull().sum()
    if missing_device > 0:
        print(f"Found {missing_device} rows with missing device_type - setting to 'unknown'")
        df_clean['device_type'] = df_clean['device_type'].fillna('unknown')
    
    print(f"Final rows after cleaning: {len(df_clean)}")
    print(f"Unique sessions: {df_clean['session_id'].nunique()}")
    
    # CRITICAL FIX: Apply session validation and outlier detection
    df_clean = detect_outliers_and_validate_sessions(df_clean)
    
    return df_clean

def analyze_user_engagement(logs_df: pd.DataFrame) -> Tuple[Dict[str, Any], pd.DataFrame]:
    """
    Analyze user engagement metrics with proper calculations and error handling.
    
    Args:
        logs_df: DataFrame with columns: session_id, duration, page_views, device_type
        
    Returns:
        Tuple of (metrics_dict, device_metrics_df)
    """
    try:
        # Validate and clean data
        df_clean = validate_data(logs_df)
        
        # FIXED: Group by session_id and aggregate properly
        session_data = df_clean.groupby('session_id').agg({
            'duration': 'sum',  # Total session duration
            'page_views': 'sum',  # Total page views per session
            'device_type': 'first'  # Device type (assuming consistent per session)
        }).reset_index()
        
        # FIXED: Calculate bounce rate correctly
        # Bounce rate = percentage of sessions with only 1 page view
        bounce_rate = (session_data['page_views'] == 1).mean()
        
        # FIXED: Calculate average session time from aggregated session data
        avg_session_time = session_data['duration'].mean()
        
        # FIXED: Calculate pages per session correctly
        pages_per_session = session_data['page_views'].mean()
        
        # Calculate additional useful metrics
        total_sessions = len(session_data)
        total_page_views = session_data['page_views'].sum()
        total_time = session_data['duration'].sum()
        
        metrics = {
            'bounce_rate': round(bounce_rate, 4),
            'avg_session_time': round(avg_session_time, 2),
            'pages_per_session': round(pages_per_session, 2),
            'total_sessions': total_sessions,
            'total_page_views': int(total_page_views),
            'total_time_minutes': round(total_time / 60, 2),
            'median_session_time': round(session_data['duration'].median(), 2),
            'session_time_std': round(session_data['duration'].std(), 2)
        }
        
        # FIXED: Device metrics using session-level data
        device_metrics = session_data.groupby('device_type').agg({
            'session_id': 'count',  # Number of sessions per device
            'duration': ['mean', 'median', 'std'],  # Session duration stats
            'page_views': ['mean', 'sum']  # Page view stats
        }).round(2)
        
        # Flatten column names
        device_metrics.columns = ['_'.join(col).strip() for col in device_metrics.columns]
        device_metrics = device_metrics.rename(columns={
            'session_id_count': 'total_sessions',
            'duration_mean': 'avg_session_time',
            'duration_median': 'median_session_time',
            'duration_std': 'session_time_std',
            'page_views_mean': 'avg_pages_per_session',
            'page_views_sum': 'total_page_views'
        })
        
        # Add device distribution percentage
        device_metrics['session_percentage'] = (
            device_metrics['total_sessions'] / total_sessions * 100
        ).round(2)
        
        print("\nEngagement Metrics Summary:")
        print(f"Bounce Rate: {metrics['bounce_rate']:.1%}")
        print(f"Average Session Time: {metrics['avg_session_time']:.1f} seconds")
        print(f"Pages per Session: {metrics['pages_per_session']:.2f}")
        print(f"Total Sessions: {metrics['total_sessions']:,}")
        
        return metrics, device_metrics
        
    except Exception as e:
        print(f"Error in analyze_user_engagement: {str(e)}")
        raise

def create_engagement_visualizations(metrics: Dict[str, Any], device_metrics: pd.DataFrame, 
                                   session_data: pd.DataFrame = None) -> None:
    """
    Create summary visualizations for engagement metrics.
    
    Args:
        metrics: Dictionary of engagement metrics
        device_metrics: DataFrame with device-specific metrics
        session_data: Optional DataFrame with session-level data for distributions
    """
    try:
        # Set up the plot style
        plt.style.use('default')
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle('User Engagement Analytics Dashboard', fontsize=16, fontweight='bold')
        
        # 1. Device Distribution Pie Chart
        ax1 = axes[0, 0]
        device_sessions = device_metrics['total_sessions']
        colors = plt.cm.Set3(np.linspace(0, 1, len(device_sessions)))
        
        wedges, texts, autotexts = ax1.pie(device_sessions.values, 
                                          labels=device_sessions.index,
                                          autopct='%1.1f%%',
                                          colors=colors,
                                          startangle=90)
        ax1.set_title('Sessions by Device Type')
        
        # 2. Average Session Time by Device
        ax2 = axes[0, 1]
        bars = ax2.bar(device_metrics.index, device_metrics['avg_session_time'], 
                       color=colors[:len(device_metrics)])
        ax2.set_title('Average Session Time by Device (seconds)')
        ax2.set_ylabel('Time (seconds)')
        ax2.tick_params(axis='x', rotation=45)
        
        # Add value labels on bars
        for bar in bars:
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                    f'{height:.1f}s', ha='center', va='bottom')
        
        # 3. Key Metrics Summary
        ax3 = axes[1, 0]
        key_metrics = {
            'Bounce Rate': f"{metrics['bounce_rate']:.1%}",
            'Avg Session Time': f"{metrics['avg_session_time']:.1f}s",
            'Pages/Session': f"{metrics['pages_per_session']:.2f}",
            'Total Sessions': f"{metrics['total_sessions']:,}"
        }
        
        y_pos = np.arange(len(key_metrics))
        colors_metrics = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
        
        # Create horizontal bar chart for key metrics
        bars = ax3.barh(y_pos, [metrics['bounce_rate'], 
                                metrics['avg_session_time']/100,  # Scale for visualization
                                metrics['pages_per_session'],
                                metrics['total_sessions']/1000],  # Scale for visualization
                       color=colors_metrics)
        
        ax3.set_yticks(y_pos)
        ax3.set_yticklabels(list(key_metrics.keys()))
        ax3.set_title('Key Engagement Metrics')
        ax3.set_xlabel('Normalized Values')
        
        # Add actual values as text
        for i, (bar, value) in enumerate(zip(bars, key_metrics.values())):
            ax3.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
                    value, ha='left', va='center', fontweight='bold')
        
        # 4. Pages per Session by Device
        ax4 = axes[1, 1]
        bars = ax4.bar(device_metrics.index, device_metrics['avg_pages_per_session'],
                       color=colors[:len(device_metrics)])
        ax4.set_title('Average Pages per Session by Device')
        ax4.set_ylabel('Pages per Session')
        ax4.tick_params(axis='x', rotation=45)
        
        # Add value labels
        for bar in bars:
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                    f'{height:.2f}', ha='center', va='bottom')
        
        plt.tight_layout()
        plt.show()
        
        # Additional distribution plots if session data is provided
        if session_data is not None:
            create_distribution_plots(session_data)
            
    except Exception as e:
        print(f"Error creating visualizations: {str(e)}")

def create_distribution_plots(session_data: pd.DataFrame) -> None:
    """Create distribution plots for session metrics with outlier handling."""
    try:
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        fig.suptitle('Session Metrics Distributions', fontsize=14, fontweight='bold')
        
        # Session duration distribution with outlier handling
        ax1 = axes[0]
        # Cap extreme values for better visualization
        duration_cap = session_data['duration'].quantile(0.99)
        duration_viz = np.minimum(session_data['duration'], duration_cap)
        
        ax1.hist(duration_viz, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
        ax1.set_title('Session Duration Distribution (99th percentile capped)')
        ax1.set_xlabel('Duration (seconds)')
        ax1.set_ylabel('Frequency')
        ax1.axvline(session_data['duration'].mean(), color='red', linestyle='--', 
                   label=f'Mean: {session_data["duration"].mean():.1f}s')
        if duration_cap < session_data['duration'].max():
            ax1.text(0.7, 0.9, f'Capped at {duration_cap:.0f}s\n({session_data["duration"].max():.0f}s max)', 
                    transform=ax1.transAxes, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
        ax1.legend()
        
        # Page views distribution with outlier handling
        ax2 = axes[1]
        # Cap extreme values for better visualization
        page_views_cap = min(session_data['page_views'].quantile(0.99), 50)  # Cap at 50 for readability
        page_views_viz = np.minimum(session_data['page_views'], page_views_cap)
        
        bins = range(1, int(page_views_viz.max()) + 2)
        ax2.hist(page_views_viz, bins=bins, alpha=0.7, color='lightgreen', edgecolor='black')
        ax2.set_title('Page Views per Session Distribution (outliers capped)')
        ax2.set_xlabel('Page Views')
        ax2.set_ylabel('Frequency')
        ax2.axvline(session_data['page_views'].mean(), color='red', linestyle='--',
                   label=f'Mean: {session_data["page_views"].mean():.2f}')
        if page_views_cap < session_data['page_views'].max():
            ax2.text(0.7, 0.9, f'Capped at {page_views_cap:.0f}\n({session_data["page_views"].max():.0f} max)', 
                    transform=ax2.transAxes, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Error creating distribution plots: {str(e)}")

# Example usage and testing function
def test_engagement_analysis():
    """Test the engagement analysis with sample data."""
    # Create sample data for testing
    np.random.seed(42)
    
    sample_data = {
        'session_id': [f'session_{i}' for i in range(1000)],
        'duration': np.random.exponential(120, 1000),  # Average 2 minutes
        'page_views': np.random.poisson(3, 1000) + 1,  # At least 1 page view
        'device_type': np.random.choice(['mobile', 'desktop', 'tablet'], 1000, p=[0.6, 0.3, 0.1])
    }
    
    # Add some edge cases for testing
    sample_data['duration'][0] = -50  # Negative duration
    sample_data['page_views'][1] = 0   # Invalid page views
    sample_data['device_type'][2] = None  # Missing device type
    
    df = pd.DataFrame(sample_data)
    
    print("Testing with sample data...")
    metrics, device_metrics = analyze_user_engagement(df)
    
    print("\nDevice Metrics:")
    print(device_metrics)
    
    # Get session data for visualizations
    df_clean = validate_data(df)
    session_data = df_clean.groupby('session_id').agg({
        'duration': 'sum',
        'page_views': 'sum',
        'device_type': 'first'
    }).reset_index()
    
    create_engagement_visualizations(metrics, device_metrics, session_data)

if __name__ == "__main__":
    test_engagement_analysis()
```
